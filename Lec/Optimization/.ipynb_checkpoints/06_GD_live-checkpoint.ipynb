# Optimization

Almost all engineering/scientific/data science problem can be transformed to an optimization problem. The history of optimization dates back to 18th-19th centuries. The root of optimization is finding the maximum or minimum value of a function. 

## Modern motivation:

**Training machine learning models is solving an optimization problem.**


Considering the simple linear regression problem where we want to find a straight line to fit given data points well, see picture:




**Fact: A straight line can be modeled as y=kx+b where k is called slope and b is called intercept.**

Our goal is to find $k$ and $b$ such that the obtained straight line fits your data well (how to define this mathematically?)


**Methodology:**

We set up the following minimization problem

$$\mathop{\mathbf{minimize}}_{k,b} \sum_{i=1}^{m} (y_i - (kx^{(i)} + b))^2, $$
where m is the number of data samples.


Above optimization problem is a simple problem in the sense that we are able to derive explicit formula for $k$ and $b$ (we will see this in the future). However, when we train neural networks, we can set up a similar optimization problem which is really hard to solve. Currently, people are using gradient descent (GD) and its variation stochastic gradient descent (SGD) to solve optimization problem.

## General mathematical formulation

Mathematically, optimization problem takes the following form

$$\mathop{\mathbf{minimize}/\mathbf{maximize}} f(x) \qquad\mbox{ subject to } \mbox{ constraints }, $$
where $f(x)$ is called **objective function** and constraints are usually inequalities or equalities. It is possible that we do not have any constraints.

## GD: Theory

1. GD/SGD is a mathematical algorithm, which is used to solve optimization problems. The implementation of GD is based on **derivative**, or more general, **gradient**. I assume that you know how to compute derivative and gradient. 



2. Things will get hard if you do not know gradient or the gradient of your objective function is hard to compute by hand. It requires researchers to design algorithms that do not depend on the gradient. The latter one always happen in the machine learning community.



3. GD is designed using the fact that **if function $f(x)$ has maximum value or minimum value, then $f'(x)=0$.** However, the converse is not true.



4. Suppose that there exists $x^*$ such that $f(x^*)\leq f(x)$ for all $x$, then our goal is to find $x^*$. It is possible that $x^*$ is not unique.



5. GD is an iterative algorithm which produces a sequence of points ${x_1,x_2,...,x_n}$ closing to our target $x^*$. The updating rule is

$$ x_{t+1} = x_{t} - \eta \nabla f(x_t),$$
where $\eta$ is called stepsize or learning rate. Here, $x_t$ is a vector. If $x_t$ is a real number, then the gradient $\nabla f(x_t)$ reduces to derivative $f'(x_t)$.



6. Gradient tells you the direction that the objective function decreases fastes. Stepsize determines how much you move each time. 



7. When points $x_t$ and $x_{t+1}$ are close, we can show that $f'(x_t)$ is close to 0 which means that we are closing to our target $x^*$. **Warning: Honestly, this is not correct in general but GD is still an acceptable algorithm.**

## Gradient descent (GD)

Now, we are ready to write our own GD algorithm.

```python
# import required packages
import numpy as np
import matplotlib.pyplot as plt
```

Our goal is to minimize $f(x)=x^2$. The minimizer is $x=0$. Now we write a gradient descent algorithm to compute it.

```python
def f(x):
    '''
    x is a real number
    '''
    return x**2

def grad_f(x):
    '''
    x is a real number
    '''
    return 2*x

def GD(x, eta, fun):
    '''
    x is a real number
    eta: stepsize, a real number
    fun: is a function, which is the derivative of your objective function
    '''
    return x - eta*fun(x)
```

```python
# initial point x0
x = 5 

# stepsize
eta = 0.05

# number of iterations
iterations = 200

for i in range(iterations):
    x = GD(x, eta, grad_f)
    if i % 20 == 0:
        print(f'After iteration {i + 1}, the updated point is x = {x}')
```

Many questions appear:

1. Can you guarantee that your algorithm converge? Even if your algorithm converges, does it converge to the desired minimizer?
2. What is the best step size? 
3. How many iterations do we need?
4. ...


Theory: GD works well for convex function in theory since the local minimizer is global minimizer. Old theory is based on the assumption that your objective function is convex. However, when training many models, in particular, deep neural networks, the objective function is proven to be non-convex. But GD and its variations are also widely used and has good performance. This motivates us to find new optimization theory without assuming that objective function is convex.

I will not prove any mathematical theory here. Let's use some examples to explain it.

#### Observation 1: A good choice of step size improves your algorithm, see next example.

```python
# initial point x0
x = 5 

# stepsize
eta = 0.05
# eta = 1

# number of iterations
iterations = 100

x_t = np.zeros(iterations + 1)

x_t[0] = x

for i in range(iterations):
    x = x_t[i]
    x_t[i + 1] = GD(x, eta, grad_f)
    if i % 20 == 0:
        print(f'After iteration {i + 1}, the updated point is x = {x_t[i + 1]}')

xx = np.linspace(-5, 5, 100)
yy = xx**2

plt.plot(xx, yy)
plt.plot(x_t, x_t**2, c = 'red', marker = 'o', markersize = 10)
```

#### Conclusion 1:

Usually, we choose small stepsize. But we do not want extremely small stepsize because smaller stepsize leads to larger iterations.

Finding a good choice of step size is left for HW. Here, we want to illustrate adaptive step size. The idea is the following:
1. When we start iterations, we can use large stepsize to make objective function decrease fast;
2. When we are close to the minimizer (after many iterations), we want small step size because we do not want our iteration move too fast.

Adaptive stepsize is designed to satisfy above requirements. There are many ways to design adaptive stepsize. Here we introduce one

## Question 2: How many iterations do we need? 

Our previous results show that the number of iterations depends on the step size. 

Theoretically, large iteration guarantees convergence but it requires more computation. Can we still implement GD without knowing number of iterations?

The answer is yes. We should change our stop criteria. For example, we can compute gradient, if it is smaller than a small threshold, then we stop.

```python
x = 5
eta = 0.02
t = 0

while abs(grad_f(x)) > 1e-6:
    t += 1
    x = GD(x, eta, grad_f)

print(f'After {t} iterations, GD converges to a  minimizer.')
print(f'Obtained minimizer is x = {x}')
```

## Observation 3: Nonconvex functions case

Things are really complicated if the objective function is nonconvex.

```python
def g(x):
    '''
    x can a numpy array
    '''

    return np.sin(np.pi*x) + 2*np.sin(2*np.pi*x)

def g_grad(x):

    return np.pi*np.cos(np.pi*x) + 4*np.pi*np.cos(2*np.pi*x)

xx = np.linspace(0, 4*np.pi, 100)

plt.plot(xx, g(xx))
```

#### Case 1: stay at the local minimum

```python
# x = 2
x = 0.5
threshold = 1e-4
eta = 0.01
x_t = x

while abs(g_grad(x_t)) > threshold:
    x_t = GD(x_t, eta, g_grad)

xx = np.linspace(0, 4*np.pi, 100)
plt.plot(xx, g(xx))
plt.scatter([x, x_t], [g(x), g(x_t)], c = ['red', 'green'])
```
