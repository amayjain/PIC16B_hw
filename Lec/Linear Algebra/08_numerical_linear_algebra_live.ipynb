## Part 2: Linear system

We will discuss solving linear system and least square problem.

In many applications, we need to solve a linear system $Ax=b$. The goal is to find $x$ given matrix $A$ and output $b$. If $A$ is a square matrix and invertible, then we can solve the problem by multiplying $A^{-1}$ for both sides. However, computing $A^{-1}$ is expensive. In python, we can use numpy.linalg.solve command to solve $Ax=b$.

```python
A = np.array([[1, 2], 
              [2, 5]])

b = np.array([[3], 
              [7]])

# solve by using inverse

x = np.linalg.inv(A) @ b

print(x, '\n')

# use solve command

x = np.linalg.solve(A, b)

print(x, '\n')


# compare
print('solve command is used to solve linear system, but it is computationally cheap')
```

Can we solve linear system $Ax=b$ when $A$ is not a square matrix or $A$ is a square matrix but not invertible?

## Least-square problem

We want to find $x$ such that $Ax$ is as close to $b$ as possible, i.e.

$$ \mathop{ \mathrm{min} }_x \|Ax-b\|_2 $$

- if we minimize L2 norm ($\|x\|_2$), it has a nice formula $x = (A^TA)^{-1}A^Tb$. This nice formula comes form normal equation $A^TAx = A^Tb$, which can be obtained by finding the gradient of objective function and setting it to zero.

Minimize L2 norm problem is also called least square problem. Linear regression can be written in this form. Actually, many problems can be written as least square problem.

We can also minimize other norms, but we do not have nice formula. But we can design other algorithms to solve it. 


```python
# verification:
A = np.array([[1, 1], 
              [2, 2], 
              [1, 0]])

b = np.array([[5], 
              [10], 
              [7]])

# np.linalg.solve(A, b)
# np.linalg.inv(A)

# inverse matrix

x_inv = np.linalg.inv(A.T @ A) @ A.T @ b


# solve command

x_solve = np.linalg.solve(A.T @ A, A.T @ b)


# compare

np.allclose(x_inv, x_solve)
```

```python
# linear regression is a least-square problem:

# Step 1
# in-class: mean-squared error can be written as a matrix form

# Step 2
# verify sklearn.linear_model.LinearRegression() is the least-square problem

# generate data points

m = 100

x = np.linspace(-10, 10, m)

y = 2*x - 3 + np.random.randn(m)

# Linear regression command

from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x.reshape(m, 1), y)

k_LR = reg.coef_[0]
b_LR = reg.intercept_

print('Linear regression command gives that')
print(f"slope is {k_LR:.12f} and intercept is {b_LR:.12f}")
print(' ')

# least-square problem

A = np.column_stack((np.ones((m, 1)), x))

coef = np.linalg.solve(A.T @ A, A.T @ y)

print('Linear regression command gives that')
print(f"slope is {coef[1]:.12f} and intercept is {coef[0]:.12f}")
```

## Part 3: Eigen decomposition and Singular value decomposition

#### Eigenvalue and eigenvector

Given a square matrix $A\in\mathbb{R}^{n\times n}$, a scalar $\lambda$ and a vector $v\in\mathbb{R}^n$ ($v\neq0$) satisfy
$$ Av = \lambda v,$$
then scalar $\lambda$ is called eigenvalue and vector $v$ is called eigenvector.

#### Eigendecomposition

For a certain class of square matrix $A\in\mathbb{R}^{n\times n}$, we can decompose it as 
$$ A = QDQ^{-1}$$
where $D$ is a diagonal matrix and $D_{i,i}$ is an eigenvalue of matrix $A$, the $i$-th column of $Q$ is an eigenvector of $A$ corresponding to eigenvalue $D_{i,i}$.

#### Singular value decomposition

The singular value decomposition of an $m\times n$ matrix $A$ is a factorization of the form

$$ M = U\Sigma V^T$$

where $U$ is an $m\times m$ unitary matrix, $\Sigma$ is an $m\times n$ rectangular diagonal matrix with non-negative real numbers on the diagonal, $V$ is an $n\times n$ unitary matrix.

**Unitary matrix:** $U$ is a unitary matrix if $U^T = U^{-1}$.
